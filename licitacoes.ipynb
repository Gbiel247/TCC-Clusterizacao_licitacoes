```python
%pip install pandas
import pandas as pd
import os
```

```python
arquivos = [
    "./2024/comprasGOV-anual-VW_DM_PNCP_ITEM_RESULTADO-2024.csv",
    "./2024/comprasGOV-anual-VW_FT_PNCP_COMPRA_ITEM-2024.csv",
    "./2024/comprasGOV-anual-VW_FT_PNCP_COMPRA-2024.csv",
    "./2025/comprasGOV-anual-VW_DM_PNCP_ITEM_RESULTADO-2025.csv",
    "./2025/comprasGOV-anual-VW_FT_PNCP_COMPRA_ITEM-2025.csv",
    "./2025/comprasGOV-anual-VW_FT_PNCP_COMPRA-2025.csv"
]

lista_compra = []
lista_item = []
lista_resultado = []

print("Iniciando a leitura e separação dos arquivos...\n")

for caminho in arquivos:
    nome_arquivo = os.path.basename(caminho)
    print(f"Lendo: {nome_arquivo}")
    
    try:
        df_temp = pd.read_csv(caminho, sep=',', encoding='utf-8')
        
        if 'COMPRA_ITEM' in nome_arquivo:
            lista_item.append(df_temp)
        elif 'ITEM_RESULTADO' in nome_arquivo:
            lista_resultado.append(df_temp)
        elif 'COMPRA-' in nome_arquivo: 
            lista_compra.append(df_temp)
            
    except Exception as e:
        print(f"Erro ao ler {nome_arquivo}: {e}")

print("\nUnificando as bases...")

df_compra_final = pd.concat(lista_compra, ignore_index=True)
df_item_final = pd.concat(lista_item, ignore_index=True)
df_resultado_final = pd.concat(lista_resultado, ignore_index=True)

print(f"Base de COMPRAS unificada com {len(df_compra_final)} linhas.")
print(f"Base de ITENS unificada com {len(df_item_final)} linhas.")
print(f"Base de RESULTADOS unificada com {len(df_resultado_final)} linhas.")
```
